```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
def entropy(p):
    return -(p * np.log2(p) + (1-p) * np.log2(1-p))
```

```python
def gini_index(p):
    return 1 - p**2 - (1-p)**2
```

```python
def misclassification(p):
    return min(p, 1-p)
```

```python
x = [i for i in np.arange(0, 1, 0.001)]
y_entropy = [entropy(p) for p in x]
y_gini = [gini_index(p) for p in x]
y_misclass = [misclassification(p) for p in x]
data = pd.DataFrame({'entropy': y_entropy,
                     'gini_index': y_gini,
                     'misclassification': y_misclass},
                   index=x)
data.head()
```

```python
fig, ax = plt.subplots(figsize=(8,8))
sns.lineplot(data=data)
```
